## üì¶ IMPORT LIBRARY STANDARD

```python
# Cell 1: Import semua library yang sudah terinstall
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Konfigurasi tampilan
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-whitegrid')
%matplotlib inline

print("‚úÖ Semua library berhasil diimport!")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

## üéØ CONTOH ANALISIS DATA CEPAT

```python
# Cell 2: Buat dataset contoh
np.random.seed(42)
data = pd.DataFrame({
    'Kategori': np.random.choice(['A', 'B', 'C', 'D'], 100),
    'Nilai_X': np.random.normal(50, 15, 100),
    'Nilai_Y': np.random.exponential(20, 100),
    'Label': np.random.choice(['Ya', 'Tidak'], 100)
})

print("üìä Dataset Contoh (5 baris pertama):")
display(data.head())

print("\nüìà Statistik Deskriptif:")
display(data.describe())
```

## üìä VISUALISASI CONTOH

```python
# Cell 3: Visualisasi
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 1. Histogram
axes[0].hist(data['Nilai_X'], bins=20, edgecolor='black', alpha=0.7)
axes[0].set_title('Distribusi Nilai X')
axes[0].set_xlabel('Nilai')
axes[0].set_ylabel('Frekuensi')

# 2. Box plot
data.boxplot(column='Nilai_Y', by='Kategori', ax=axes[1])
axes[1].set_title('Nilai Y per Kategori')
axes[1].set_xlabel('Kategori')
axes[1].set_ylabel('Nilai Y')

# 3. Scatter plot
scatter = axes[2].scatter(data['Nilai_X'], data['Nilai_Y'], 
                         c=data['Label'].map({'Ya': 'green', 'Tidak': 'red'}), 
                         alpha=0.6)
axes[2].set_title('Hubungan X vs Y')
axes[2].set_xlabel('Nilai X')
axes[2].set_ylabel('Nilai Y')

# Legend
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=10, label='Ya'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=10, label='Tidak')
]
axes[2].legend(handles=legend_elements)

plt.tight_layout()
plt.show()

print("üé® Visualisasi selesai!")
```

## üíæ CARA SIMPAN FILE

```python
# Cell 4: Fungsi untuk simpan dan download
def daftar_file():
    """Tampilkan semua file di direktori"""
    import os
    files = os.listdir('.')
    print("üìÇ File di direktori saat ini:")
    for f in sorted(files):
        if not f.startswith('.'):  # Skip hidden files
            print(f"  ‚Ä¢ {f}")
    
    # Buat file contoh untuk di-download
    data.to_csv('contoh_data.csv', index=False)
    print("\n‚úÖ File 'contoh_data.csv' sudah dibuat!")
    print("üí° Untuk download: Klik kanan file ‚Üí Download")

# Jalankan fungsi
daftar_file()
```
